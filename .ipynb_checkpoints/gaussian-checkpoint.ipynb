```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
stdev = 30
mean = 100
```

```python
def gaussian(x, mean, stdev):
    a = 1 / (stdev * np.sqrt(2 * np.pi))
    b = np.exp(-1 * np.square(x - mean) / (2 * stdev * stdev))
    return a * b
```

```python
r = 200
i = 1

xaxis = []

while i <= r:
    xaxis.append(i)
    i += 1
    
yaxis = []
for i in xaxis:
    op = gaussian(i, mean, stdev)
    yaxis.append(op)
    
plt.plot(xaxis, yaxis)
plt.show()
```
